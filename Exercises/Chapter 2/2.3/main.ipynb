#### Q10

```python
A = torch.randn(1<<10, 1<<16)
B = torch.randn(1<<16, 1<<5)
C = torch.randn(1<<5, 1<<14)
```

```python
A @ (B @ C)  # 2^35 + 2^40 (Fun fact: runtime 3.7s in cpu, much faster than I had imagined.)
```

```python
(A @ B) @ C  # 2^31 + 2^29
```
